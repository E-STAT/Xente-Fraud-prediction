```python
################################ Time to use Label encoding
from sklearn.preprocessing import LabelEncoder
encoder = LabelEncoder()
cat_features = ['BatchId', 'ProviderId', 'ProductId',
       'ProductCategory', 'ChannelId', 'Value', 'PricingStrategy',
       'Batch_TransactionId', 'Batch_AccountId', 'Batch_SubscriptionId',
       'Batch_CustomerId', 'Provider_TransactionId', 'Provider_AccountId',
       'Provider_SubscriptionId', 'Provider_CustomerId',
       'Product_TransactionId', 'Product_AccountId', 'Product_SubscriptionId',
       'Product_CustomerId', 'Product_Cat_TransactionId',
       'Product_Cat_AccountId', 'Product_Cat_SubscriptionId',
       'Product_Cat_CustomerId', 'Channel_TransactionId', 'Channel_AccountId',
       'Channel_SubscriptionId', 'Channel_CustomerId', 'Product_Cat_ChannelId',
       'Provide_ProductId']

encoder = LabelEncoder()
encoded = Xente_data[cat_features].apply(encoder.fit_transform)
```

```python
Xente_data.drop(['BatchId', 'ProviderId', 'ProductId',
       'ProductCategory', 'ChannelId', 'Value', 'PricingStrategy',
       'Batch_TransactionId', 'Batch_AccountId', 'Batch_SubscriptionId',
       'Batch_CustomerId', 'Provider_TransactionId', 'Provider_AccountId',
       'Provider_SubscriptionId', 'Provider_CustomerId',
       'Product_TransactionId', 'Product_AccountId', 'Product_SubscriptionId',
       'Product_CustomerId', 'Product_Cat_TransactionId',
       'Product_Cat_AccountId', 'Product_Cat_SubscriptionId',
       'Product_Cat_CustomerId', 'Channel_TransactionId', 'Channel_AccountId',
       'Channel_SubscriptionId', 'Channel_CustomerId', 'Product_Cat_ChannelId',
       'Provide_ProductId'], axis = 1, inplace = True)
```

```python
Xente_data = Xente_data.join(encoded)
```

```python
Xente_data.head(20)
```

```python
y = Xente_data['FraudResult']
X = Xente_data.drop(['FraudResult'], axis=1)
```

```python
import lightgbm as lgb
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.metrics import auc, accuracy_score, roc_auc_score, roc_curve, f1_score
from sklearn.model_selection import GridSearchCV
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.30, random_state=0)
X_train.columns
```

```python
estimator = lgb.LGBMClassifier(learning_rate = 0.15, metric = 'l1', 
                        n_estimators = 24, num_leaves = 38)


param_grid = {
    'n_estimators': [x for x in range(24,40,2)],
    'learning_rate': [0.10, 0.125, 0.15, 0.175, 0.2]}
gridsearch = GridSearchCV(estimator, param_grid)

gridsearch.fit(X_train, y_train,
        eval_set = [(X_test, y_test)],
        eval_metric = ['auc', 'binary_logloss', 'f1_score'],
        early_stopping_rounds = 5,
              verbose = 1)
```

```python
print('Best parameters found by grid search are:', gridsearch.best_params_)
```

```python
gbm = lgb.LGBMClassifier(learning_rate = 0.125, metric = 'l1', 
                        n_estimators = 24)


gbm.fit(X_train, y_train,
        eval_set=[(X_test, y_test)],
        eval_metric=['auc', 'binary_logloss', 'f1_score'],
early_stopping_rounds=5)
```

```python
gbm.score(X_test, y_test)
```

```python
gbm.evals_result_
```

```python
y_pred = gbm.predict(X_test)
y_pred
```

```python
evaluation = f1_score(y_pred, y_test)
evaluation
```
